```python
import pandas as pd

# Load the dataset
file_path = r"C:\Users\Maitri Chitania\OneDrive\Desktop\Technical\AICTE INTERSHIP\Datasets\heart_disease_data.xlsx"
df = pd.read_excel(file_path, engine="openpyxl")  # Make sure this loads correctly

# Check available columns
print(df.columns)

# Define Features (X) and Target (Y)
X = df.drop(columns=['LUNG_CANCER'], axis=1)  # Drop the target column from features
Y = df['LUNG_CANCER']  # Target variable

# Verify the data
print(X.head())
print(Y.head())
```

```python
print(df.columns)
```

```python
# Define Features (X) and Target (Y)
X = df.drop(columns=['target'], axis=1)  # Drop the target column from features
Y = df['target']  # Target variable

# Verify the data
print(X.head())  # Check features
print(Y.head())  # Check target
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train.values, Y_train.values)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```
